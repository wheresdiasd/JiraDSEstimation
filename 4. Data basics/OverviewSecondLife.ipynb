```python
from pymongo import MongoClient
import pandas as pd
import numpy as np

# Connect to MongoDB
client = MongoClient("mongodb://admin:password@localhost:27017/")
db = client["JiraRepos"]

collection = db["Hyperledger"]
issues = list(collection.find())

if issues:
    df = pd.DataFrame(issues)
    print(f"Loaded {len(df)} issues from Second Life repository.")
else:
    print("No issues found in Second Life repository.")


# Check number of documents
num_records = collection.count_documents({})
print(f"‚úÖ Number of records: {num_records}")

# Fetch a sample of 5 records
sample_data = list(collection.find().limit(5))
df_sample = pd.DataFrame(sample_data)

# Display columns
print("üõ†Ô∏è Available Columns:")
print(df_sample.columns)

# Show a preview of the data
df_sample.head()


# Extract key details from the dataset
df_expanded = df_sample.copy()

# Expand the 'fields' dictionary to get issue details
df_expanded = pd.concat([df_expanded.drop(['fields'], axis=1), df_expanded['fields'].apply(pd.Series)], axis=1)

# Show available fields inside 'fields'
print("üõ†Ô∏è Available fields in 'fields' column:")
print(df_expanded.columns)

# Show a preview of the expanded dataset
df_expanded.head()
```
